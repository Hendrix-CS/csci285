```python
class RANDU():
    
    def __init__(self, s=1):
        self.SEED = s
        
    def random(self):
        self.SEED *= 65539
        self.SEED %= 2 ** 31
        return self.SEED / 2 ** 31
```

```python
r = RANDU(99)
for i in range(10):
    print(r.random())
```

```python
nums2 = []
for i in range(254):
    nums2.append(1 + int(r.random()*10))
```

```python
df["RANDU"] = nums2
```

```python
ggplot(df, aes(x="RANDU")) + geom_histogram(bins=10)
```

```python
counts, bins = np.histogram(df["RANDU"], bins=np.arange(1, 12))
```

# Python Random 

https://docs.python.org/3/library/random.html

```python
import random
r = random.Random(99)
```

```python
points = []
for i in range(10000):
    points.append((r.random(), r.random(), r.random()))
df3d = pd.DataFrame(points, columns=["x", "y", "z"])
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection = "3d")
ax.view_init(elev=20., azim = 60)
ax.scatter(df3d["x"], df3d["y"], df3d["z"], s=1)
```
