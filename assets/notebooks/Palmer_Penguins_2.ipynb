## Visualizing Data with Seaborn

```python
# Import pandas
import pandas as pd
pd.set_option('display.max_columns', None)

# Import seaborn
import seaborn as sns
%matplotlib inline

# Apply the default theme
sns.set_theme()
```

### Load palmer penguins dataset into a `pandas.DataFrame`

```python
df = pd.read_csv('../data/penguins.csv')
```

### Histogram of flipper_length_mm

```python
hist = sns.displot(
    data=df, 
    x="flipper_length_mm", 
    kind="hist", 
    aspect=1.5
).set(
    title="Flipper Length (mm)"
)
```

### Histogram of flipper_length_mm (grouped by species)

```python
hist = sns.displot(
    data=df, 
    x="flipper_length_mm", 
    kind="hist", 
    aspect=1.5,
    hue="species", 
    multiple="stack"
).set(
    title="Flipper Length (mm)"
)
```

### Histogram of bill_length_mm (grouped by species)

```python
hist = sns.displot(
    data=df, 
    x="bill_length_mm", 
    kind="hist", 
    aspect=1.5, 
    hue="species", 
    multiple="dodge"
).set(
    title = "Bill Length (mm)"
)
```

### Histogram of bill_depth_mm (grouped by species)

```python
kde = sns.displot(
    data=df, 
    x="bill_depth_mm", 
    kind="kde", 
    aspect=1.5, 
    hue="species", 
    multiple="stack"
).set(
    title="Bill Depth (mm)"
)
```

### Scatterplot of  flipper_length_mm vs. bill_length_mm (grouped by species)

```python
scatter = sns.relplot(
    data=df, 
    x="flipper_length_mm", 
    y="bill_length_mm", 
    kind="scatter",
    height=7,
    aspect=1.5, 
    hue="species",
    style="species",
).set(
    title="Flipper Length (mm) vs Bill Length (mm)"
)
```

```python
scatter = sns.lmplot(
    data=df,
    x="flipper_length_mm", 
    y="bill_length_mm", 
    height=7,
    aspect=1.5, 
    hue="species",
).set(
    title="Flipper Length (mm) vs Bill Length (mm)",
)
```

### Boxplot of body_mass_g vs. species vs. sex

```python
box = sns.catplot(
    data=df,
    kind="box",
    x="sex",
    col="species",
    y="body_mass_g",
    palette="rocket",
)
```

### Swarm & violin plots of body_mass_g vs. species vs. sex

```python
swarm = sns.catplot(
    data=df,
    kind="swarm",
    x="species",
    hue="sex",
    y="body_mass_g",
    aspect=1.5,
    palette="rocket",
).set(title="Body Mass (g)")

violin = sns.catplot(
    data=df,
    kind="violin",
    x="species",
    hue="sex",
    y="body_mass_g",
    aspect=1.5,
    split=True,
    palette="rocket",
).set(title="Body Mass (g)")
```

### Pair plotting (aka "everything vs. everything")

```python
_ = sns.pairplot(
    data=df[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g', 'species']],
    hue="species",
    markers=["o", "s", "D"],
    corner=True
)
```

```python
_ = sns.pairplot(
    data=df[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g', 'sex']],
    hue="sex",
    markers=["o", "s"],
    corner=True,
    palette="rocket",
)
```
