## Estimating $\pi$

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

$$
\frac{\text{area of circle}}{\text{area of square}} = \frac{\pi \cdot r^2}{(2 \cdot r) \cdot (2 \cdot r)} = \frac{\pi \cdot r^2}{ 4 \cdot r^2} = \frac{\pi}{4}
$$

```python
num_points = 100_000
```

```python
grid = []

for point in range(num_points):
    
    rand_x = 2 * np.random.random() - 1
    rand_y = 2 * np.random.random() - 1
    
    circle = 0
    if np.sqrt(rand_x**2 + rand_y**2) <= 1:
        circle = 1
    
    grid.append([rand_x, rand_y, circle])
```

```python
df = pd.DataFrame(grid, columns=['x', 'y', 'circle'])
```

```python
df.shape
```

```python
scatter = sns.relplot(
    data=df,
    x='x',
    y='y',
    hue='circle',
    kind='scatter',
    height=7
)
```
