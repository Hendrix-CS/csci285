```python
penguin_df = pd.read_csv("data/penguins_raw.csv")
```

```python
penguin_df
```

```python
ggplot(penguin_df, aes(x="Sex", y="Body Mass (g)")) \
+ geom_boxplot()
```

```python
penguin_df = penguin_df.dropna(subset=["Sex"])
```

```python
penguin_df = penguin_df.loc[(penguin_df["Sex"] == "MALE") | (penguin_df["Sex"] == "FEMALE")]
```

```python
ggplot(penguin_df, aes(x="Culmen Length (mm)", fill="Sex")) \
+ geom_histogram(bins = 30, position="identity", alpha=0.5)
```

```python
ggplot(penguin_df, aes(y="Culmen Length (mm)", x="Flipper Length (mm)", color="Species")) \
+ geom_point() \
+ geom_smooth(method="lm") \
+ labs(title="Palmer Penguins Flipper and Bill Relationship")
```
