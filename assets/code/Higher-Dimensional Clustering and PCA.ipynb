```python
%matplotlib inline
import pandas as pd
from plotnine import *
```

```python
df = pd.read_csv("data/penguins_raw.csv")
```

```python
df
```

```python
features = ['Culmen Length (mm)',
            'Culmen Depth (mm)',
            'Flipper Length (mm)',
            'Body Mass (g)']
df = df.dropna(subset=features).reset_index()
```

```python
df
```

```python
x = df.loc[:, features].values
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler
x = StandardScaler().fit_transform(x)
```

```python
x
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters = 3, init="random",
           n_init = 10, max_iter=300,
           tol=1e-04, random_state=0)
result = km.fit(x)
```

```python
result.cluster_centers_
```

```python
result.labels_
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(x)
```

```python
pcadf = pd.DataFrame(data= principal_components,
                       columns=["PCA 1", "PCA 2"])
```

```python
pcadf
```

```python
ggplot(pcadf, aes(x="PCA 1", y="PCA 2", color="cluster")) + \
geom_point()
```

```python
pcadf["cluster"] = pd.Series(result.labels_).astype("str")
```

```python
pcadf["species"] = df["Species"]
```

```python
dist = []
for i in range(1, 5):
    print(i)
    km = KMeans(
        n_clusters=i, init='random',
        n_init=10, max_iter=300, 
        tol=1e-04, random_state=0
    )
    result = km.fit(x)
    dist.append(result.inertia_)
```

```python
idf = pd.DataFrame({"Number of Clusters":range(1, 5), "Distortion":dist})
```

```python
ggplot(idf, aes(x="Number of Clusters", y="Distortion")) + geom_line() + geom_point()
```
